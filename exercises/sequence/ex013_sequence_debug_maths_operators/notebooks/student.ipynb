## Exercise 10 — Mixed challenge

**Expected output:**
```

```

````python
items = input("How many items? ")
box_size = input("How many per box? ")
total_cost = float(input("Total cost? £"))

boxes = items // box_size
leftover = items % box_size
cost_per_box = total_cost / boxes
cost_per_box = round(cost_per_box, 2)

print(f"Full boxes: {boxes}")
print(f"Leftover items: {leftover}")
print(f"Cost per box: £{cost_per_box}")
````
