## Exercise 10 — Mixed challenge

**Expected output:**
```
How many items? [⌨️: 29]
How many per box? [⌨️: 6]
Total cost? £ [⌨️: 10.99]
Full boxes: 4
Leftover items: 5
Cost per box: £2.75
```

````python
items = input("How many items? ")
box_size = input("How many per box? ")
total_cost = input("Total cost? £")

items = int(items)
box_size = int(box_size)
total_cost = float(total_cost)

boxes = items // box_size
leftover = items % box_size
cost_per_box = total_cost / boxes
cost_per_box = round(cost_per_box, 2)

print(f"Full boxes: {boxes}")
print(f"Leftover items: {leftover}")
print(f"Cost per box: £{cost_per_box}")
````

### What actually happened
Describe what errors or wrong output you got, and what you changed to fix them.

The variables were in the data form of a string, making them unable to perform arithmetic. To fix this, I converted the variables into intergers and floats.
